```python
import pandas as pd
```

```python
# Récupération du fichier .csv
df_csv = pd.read_csv('data/pokemon_data.csv')
```

```python
"""Recours à une condition
Pour les Pokemon de type 'Fire' afficher 'True' dans
champ 'Legendary'"""
df_csv.loc[df_csv['Type 1'] == 'Fire', 'Legendary'] = True
```

```python
# Récupération du fichier modified.csv
df_csv = pd.read_csv('data/modified.csv')
```

```python
df_csv.head(5)
```

```python
"""Recours à 2 conditions
Pour les Pokemon avec un total > 500 remplacement des
champs 'Generation' et 'Legendary' par 'TEST VALUE'"""
df_csv.loc[df_csv['Total'] > 500, ['Generation', 'Legendary']] = 'TEST VALUE'
```

```python
df_csv
```

```python
"""Recours à 2 conditions
Pour les Pokemon avec un total > 500 remplacement des
champs 'Generation' et 'Legendary' par :
-> 'Test 1' pour le champ 'Generation'
-> 'Test 2' pour le champ 'Legendary"""
df_csv.loc[df_csv['Total'] > 500, ['Generation', 'Legendary']] = ['Test 1', 'Test 2']
```
