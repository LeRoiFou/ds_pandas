```python
"""
Lien : https://www.youtube.com/watch?v=xPPs59pn6qU
Cours : Python Pandas Tutorial 10. Pivot table

La fonction native pivot() est assimilée à un TCD qui réorganise un tableau qui
peut avoir une double entrée d'index ou une double entrée de colonnes

La fonction native pivot_table() est également assimilée à un TCD mais elle est
utilisée en présence de doublons à la différence de la fonction native pivot()
On a en complément de la fonction pivot_table, les fonctions :
-> aggfunct
-> margins
-> grouper

Éditeur : Laurent REYNAUD
Date : 17-07-2021
"""
```

```python
import pandas as pd
```

```python
"""Chargement du fichier csv :
Tableau avec des données sur la température et l'humidité 
le 5 de chaque mois dans plusieurs villes"""

df = pd.read_csv('data/weather.csv')
df
```

```python
"Affichage des données le 5 de chaque mois dans chaque ville"

df.pivot(index='date', columns='city')
```

```python
"Même données que précédemment mais uniquement sur l'humidité"

df.pivot(index='date', columns='city', values='humidity')
```

```python
"""Chargement du fichier csv :
À la différence du précédent fichier, cette fois-ci nous avons
également les données de température et d'humidité le 5 de chaque mois
dans plusieurs villes mais les données ont été effectuées 2 FOIS dans
la même journée"""

df = pd.read_csv('data/weather2.csv')
df
```

```python
"Moyenne par jour de l'humidité et de la température par ville"

df.pivot_table(index='city', columns='date')
```

```python
"Somme par jour de l'humidité et de la température par ville"

# La fonction aggfunc est par défaut égale à 'mean' car elle 
# n'a pas été évoquée dans l'instruction précédente
df.pivot_table(index='city', columns='date', aggfunc='sum')
```

```python
"Nombre de données par jour sur l'humidité et la température par ville"

df.pivot_table(index='city', columns='date', aggfunc='count')
```

```python
"""Moyenne par jour de l'humidité et de la température par ville
avec en complément une moyenne générale"""

df.pivot_table(index='city', columns='date', margins=True)
```

```python
"""Chargement du fichier csv :
Tableau avec des données sur la température et l'humidité 
sur différents jours dans plusieurs villes"""

df = pd.read_csv('data/weather3.csv')
df
```

```python
"Type pour l'objet 'date' : str"

df.dtypes
```

```python
"Conversion de l'objet date en datetime"

df['date'] = pd.to_datetime(df['date'])

```

```python
"Moyenne par mois de l'humidité et de la température par ville"

df.pivot_table(index=pd.Grouper(freq='M', key='date'), columns='city')
```
