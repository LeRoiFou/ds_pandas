```python
import pandas as pd
```

```python
"Chargement du fichier titanic.csv"

titanic = pd.read_csv("data/titanic.csv")
titanic.head(5)
```

```python
"Tri d'un tableau par ordre croissant avec sort_values()"
# Tri du fichier titanic par age

titanic.sort_values(by='Age').head(10)
```

```python
"Tri d'un tableau par ordre décroissant avec sort_values() et ascending=False"
# Tri du fichier titanic par classe et par âge par ordre décroissant

titanic.sort_values(by=['Pclass', 'Age'], ascending=False).head(10)
```

```python
"Filtre des données à partir d'une colonne"
# Liste des passagers survivants du titanic

titanic_survived = titanic[titanic['Survived'] == 1 ]
titanic_survived.head(10)
```

```python
"Affichage par groupe avec l'instruction groupby([])"
# Affichage des 10 premiers passagers survivants par sexe

titanic_survived_sex = titanic_survived.groupby(['Sex']).head(10)
titanic_survived_sex
```

```python
"Instruction pivot() pour remodéliser le tableau avec une valeur"
# Affichage des noms des passagers n'ayant pas survécu / ayant survécu
titanic_passager = titanic.pivot(columns='Survived', values='Name')
titanic_passager.head(10)
```

```python
"Instruction pivot_table() pour remodéliser le tableau avec plusieurs valeurs = TABLEAU CROISE DYNAMIQUE"
# Moyenne des âges des passagers par sexe ayant survécu ou non au titanic

titanic_stat_sex = titanic.pivot_table(
    columns='Survived',
    index='Sex',
    values='Age',
  margins=True # Rajout index / colonne 'All'
)
titanic_stat_sex
```

```python
"Remodélisation du tableau pivot avec l'instruction melt()"

titanic_melt = pd.melt(titanic_passager)
titanic_melt.head(10)
```
