```python
import seaborn as sns
import pandas as pd
```

```python
"""Chargement d'un fichier avec en arguments :
-> index_col=0 : absence de colonne avec un index
-> parse_dates=True : ???"""

air_quality = pd.read_csv('data/air_quality_no2.csv', index_col=0, parse_dates=True)
air_quality
```

```python
"Rajout d'une colonne"
air_quality["london_mg_per_cubic"] = air_quality["station_london"] * 1.882
air_quality.head(8)
```

```python
"Renommage de colonne"
air_quality_renamed = air_quality.rename(columns={
    "station_paris": "Station de Paris",
    "station_london": "Station de Londres"})
air_quality_renamed.head(5)
```

```python
"Noms des colonnes en majuscules"
air_quality_renamed = air_quality_renamed.rename(columns=str.upper)
air_quality_renamed.head(5)
```
