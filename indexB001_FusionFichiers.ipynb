```python
import pandas as pd
import os # récupération des fichiers d'un répertoire
```

```python
"Fusion des 12 fichiers Sales_xxx en un seul fichier"

all_months_data = pd.DataFrame()

# Récupération des 12 fichiers dans le sous-répertoire 'sales' du répertoire 'data'
files = [file for file in os.listdir('data/sales')]

# Concaténation des fichiers dans le sous-répertoire 'sales'
for file in files:
    df = pd.read_csv('data/sales/'+file)
    all_months_data = pd.concat([all_months_data, df])

all_months_data.head(8)
```

```python
"Exportation de toutes les données dans un fichier CSV"

all_months_data.to_csv('data/sales/all_data.csv', index=False)
```
