##### <font color = 'green'>Intérêt de la dataframe avec Pandas </font>

À la différence de Numpy, la DF de Pandas permet d'obtenir plusieurs types de données (réels, entiers, chaînes de caractères). <br>
Mais chaque colonne n'a qu'un seul type d'objet. <br>
Le package Numpy a cependant un intérêt car certaines méthodes s'appliquent avec cette bibliothèque. <br>
À la différence de Numpy dont les arrays peuvent être multi-dimensionnels, les DF de Pandays ne sont que bi-dimensionnels. <br>
D'autre part, l'accès à un élément d'un array Numpy ne peut se faire que par des positions entières à la différence de Pandas, <br>
et les en-têtes des colonnes n'apparaissent pas dans un array numpy <br><br>

Les DF qui ressortent avec une colonne ne sont pas véritablement des DF mais des Series : les Series n'affichent pas l'en-tête des colonnes. <br>
Dans Pandas, il y a donc deux objets : les Series et les Dataframes. <br>
Or, certaines méthodes dans Pandas ne s'appliquent pour l'objet Series<br><br>

Les arrays Numpy sont spécifiquement utilisés lorsque des calculs scientifiques complexes doivent être effectués, <br>
tandis que les DataFrames sont principalement utilisés dans le prétraitement des données. <br>

##### <font color = 'yellow'>Les arrays Numpy sont similaires aux listes alors que les dataframes Pandas sont similaires aux dictionnaires</font>

##### <font color = 'green'>Cas pratique</font>

Nous allons travailler à présent sur notre fichier de prêts immobiliers. <br>

Pour expliquer rapidement ce fichier, chaque ligne correspond à un prêt qui a été accordé à un de nos clients. Chaque client est identifié par… son identifiant ! Nous avons les informations suivantes : <br>

la ville et le code postal de l’agence où le client a contracté le prêt<br>
le revenu mensuel du client<br>
les mensualités remboursées par le client<br>
la durée du prêt contracté, en nombre de mois<br>
le type de prêt<br>
et enfin le taux d’intérêt<br><br>
Votre rôle cette fois ci va être de modifier ce jeu de données pour calculer différentes variables nécessaires pour identifier les clients qui sont à la limite de leur capacité de remboursement et déterminer les bénéfices réalisées par la banque.

```python
# Importation du fichier 
prets = pd.read_csv('https://raw.githubusercontent.com/benjaminmrl/data-4452741/main/prets.csv')
prets.head()
```

Dans un premier temps, créez une nouvelle variable calculant le taux d'endettement de chaque individu. Ce taux correspond au pourcentage du revenu remboursé chaque mois par un individu. Vous arrondirez le résultat à 2 chiffres après la virgule :

```python
prets['Taux_endettement'] = round(
    prets['remboursement'] / prets['revenu'] * 100, 2)
```

Pour éviter toute confusion, renommez la variable taux en taux_interet :

```python
prets.rename(columns={'taux':'taux_interet'}, inplace=True)
```

```python
prets.head()
```

Nous allons à présent créer deux dernières variables : <br>
 
-> cout_total correspondant au cout total du pret à partir du remboursement et de la duree <br>
-> benefices correspondant aux bénéfices mensuels réalisés par la banque sur le prêt <br>

Nous simplifierons ici le calcul des bénéfices par :

bénéfices =  (C∗T) / 24 

avec :

C = cout total du prêt
T = taux d'intérêt

```python
prets['cout_total'] = prets['remboursement'] * prets['duree']
prets['benefices'] = round((
    prets['cout_total'] * (prets['taux_interet'] / 100)) / 24, 2)
```

A présent, nous souhaitons connaitre les 5 prêts sur lesquels nous produisons le plus de bénéfices. Mettez en place une solution pour les afficher :

```python
prets.sort_values(by='benefices', ascending=False).head()
```

##### <font color = 'green'>Cas pratique</font>

Nous allons à présent aller un peu plus loin dans la manipulation de nos prêts bancaires. Les demandes suivantes ont été formulées par le responsable du service prêt de notre établissement : <br>

Le taux d’endettement autorisé est de 35%. Pourriez vous me communiquer le nombre de personnes ayant dépassé ce seuil ? <br>
Même question, mais cette fois ci uniquement sur l’agence Parisienne <br>
Pour faciliter le traitement d’éventuelles futures demandes de prêts, pourriez vous ajouter une variable nommée risque qui nous permettrait d’identifier les clients risqués facilement <br>
Combien de prêts automobiles ont été accordés ? Quel est le coût total moyen de ces derniers ? <br>
Quel est le bénéfice mensuel total réalisé par l’agence Toulousaine ? <br> <br>
Vous allez devoir mettre en application l’ensemble des processus de sélection présentés ci-dessus pour répondre à ces différentes demandes.

```python
# traitement réalisés précédemment
prets = pd.read_csv('https://raw.githubusercontent.com/benjaminmrl/data-4452741/main/prets.csv')

# calcul du taux d'endettement
prets['taux_endettement'] = round(prets['remboursement'] * 100 / prets['revenu'], 2)

# renommer taux en taux_interet
prets.rename(columns={'taux':'taux_interet'}, inplace=True)

# calculer le cout total du pret
prets['cout_total'] = prets['remboursement'] * prets['duree']

# calculer les bénéfices mensuels réalisés
prets['benefices'] = round((prets['cout_total'] * prets['taux_interet']/100)/(24), 2)

prets.head()
```

Le taux d’endettement autorisé est de 35%. Pourriez vous me communiquer le nombre de personnes ayant dépassé ce seuil ? vous pourrez me stocker la liste dans une variable à part nommé clients_risque

```python
clients_risque = prets.loc[prets['taux_endettement'] > 35].shape[0]
print(f"Il y a {clients_risque} clients concernés")
```

Même question, mais cette fois ci uniquement sur l’agence Parisienne

```python
clients_risque = prets.loc[
    (prets['taux_endettement'] > 35) & 
    (prets['ville'] == 'PARIS')].shape[0]
print(f"Il y a {clients_risque} clients concernés")
```

Pour faciliter le traitement d’éventuelles futures demandes de prêts, pourriez vous ajouter une variable nommée risque qui aurait pour valeur Oui si le client est risqué (taux d’endettement > 35%), Non sinon. Pour se faire, je vous conseille de créer la variable risque en l'initialisant à Non (ou Oui, à votre préférence) et de ne modifier que les lignes concernées par la valeur inverse :

```python
prets_risque = 'Non'
prets.loc[prets['taux_endettement'] > 35, 'risque'] = 'Oui'
prets.head(10)
```

Combien de prêts automobiles ont été accordés ? Quel est le coût total moyen de ces derniers ?

```python
# Prêts automobiles accordés et coût moyen
prets_auto = prets.loc[prets['type'] == 'automobile'].shape[0]
cout_moyen = prets.loc[prets['type'] == 'automobile', 'cout_total'].mean()
print(f"{prets_auto} prêts automobiles ont été accordés pour un coût moyen de {cout_moyen} €")
```

Quel est le bénéfice mensuel total réalisé par l’agence Toulousaine ?

```python
prets_toulouse = prets.loc[prets['ville'] == 'TOULOUSE', 'benefices'].sum()
print(f"Le bénéfice mensuel total réalisé par l'agence toulousaine est de {round(prets_toulouse, 2)} €")
```
