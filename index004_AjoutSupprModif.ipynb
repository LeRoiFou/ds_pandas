```python
import pandas as pd
```

```python
# Récupération du fichier .csv
df_csv = pd.read_csv('data/pokemon_data.csv')
```

```python
# Ajout d'une colonne 'Total'
df_csv['Total'] = (df_csv['HP'] + df_csv['Attack'] + df_csv['Defense'] 
                   + df_csv['Sp. Atk'] + df_csv['Sp. Def'] + df_csv['Speed'])
```

```python
df_csv.head(5)
```

```python
# Suppression de la colonne 'Total' : 
# réinitialisation de la variable df_csv
df_csv = df_csv.drop(columns=['Total'])
```

```python
df_csv.head(5)
```

```python
# Total de la colonne indice n° 4 à l'indice n° 9 à l'horizontal
df_csv['Total'] = df_csv.iloc[:, 4:10].sum(axis=1) # axis=1 -> horizontal
```
